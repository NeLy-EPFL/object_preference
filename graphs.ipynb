```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Load the data
df = pd.read_csv('flydata_AS_resume.csv')
```

These initial plots display the raw data for the three habituation conditions, showing the total time spent with the orange or blue object, regardless of the feeding state. A boxplot accompanies the raw data, providing a visual summary of the data distribution. The table below shows the median value for each condition.

```python
# Define the color mapping
color_dict = {'orange': 'darkorange', 'blue': 'royalblue', 'unused': 'gray'}

# Create a figure with two subplots (one row, two columns)
fig, axs = plt.subplots(1, 2, figsize=(20, 10))

# Add a title to the figure
#fig.suptitle('Total time on food-objects under different habituation conditions', fontsize=18, weight='bold')

# Plot the scatterplot and boxplot for 'tot_time_blue' on the first subplot
sns.scatterplot(data=df, x='tot_time_blue', y='habituation_cond', hue='habituation_cond', palette=color_dict, legend=False, marker='x', linewidth=3, s=100, ax=axs[0])
sns.boxplot(data=df, x='tot_time_blue', y='habituation_cond', color='white', linewidth=4, fliersize=0, boxprops=dict(alpha=0.5), width=0.4, ax=axs[0])
axs[0].set_xlabel('Total Time on Blue [s]', fontsize=16, weight='bold')
axs[0].set_ylabel('Habituation Condition', fontsize=18, weight='bold')
axs[0].set_xlim([0, 3500])
axs[0].tick_params(axis='both', labelsize=16, rotation=45) 
axs[0].set_yticks([0, 1, 2])  
axs[0].set_yticklabels(['Control group', 'Orange', 'Blue'])

# Plot the scatterplot and boxplot for 'tot_time_orange' on the second subplot
sns.scatterplot(data=df, x='tot_time_oran', y='habituation_cond', hue='habituation_cond', palette=color_dict, legend=False, marker='x',linewidth=3, s=100, ax=axs[1])
sns.boxplot(data=df, x='tot_time_oran', y='habituation_cond', color='white', linewidth=3, fliersize=0, boxprops=dict(alpha=0.5), width=0.4, ax=axs[1])
axs[1].set_xlabel('Total Time on Orange [s]', fontsize=16, weight='bold')
axs[1].set_ylabel('Habituation Condition', fontsize=18, weight='bold')
axs[1].set_xlim([0, 3500])
axs[1].tick_params(axis='both', labelsize=16, rotation=45)
axs[1].set_yticks([0, 1, 2])  
axs[1].set_yticklabels(['Control group', 'Orange', 'Blue']) 
axs[1].set_ylabel('')


# Show the plot
plt.tight_layout()
plt.subplots_adjust(top=0.92)  # Adjust the top padding to make room for the title
plt.show()
```

```python
# Calculate the median of 'tot_time_blue' for each 'habituation_cond'
medians = df.groupby('habituation_cond')['tot_time_blue'].median()

# Convert the Series to a DataFrame and reset the index
medians_df = medians.reset_index()

# Rename the columns
medians_df.columns = ['Habituation Condition', 'Median Total Time on Blue']

# Display the DataFrame
medians_df
```

```python
# Calculate the median of 'tot_time_oran' for each 'habituation_cond'
medians = df.groupby('habituation_cond')['tot_time_oran'].median()

# Convert the Series to a DataFrame and reset the index
medians_df = medians.reset_index()

# Rename the columns
medians_df.columns = ['Habituation Condition', 'Median Total Time on Orange']

# Display the DataFrame
medians_df
```

This barplot shows the percentage of time each fly spent with each object and in the neutral area. We did this to get a quick idea of the general trend.

```python
# Calculate the total time for each row
df['total_time'] = df['tot_time_oran'] + df['tot_time_blue'] + df['tot_time_neutral']

# Calculate the percentage of time spent on each type of object
df['percent_oran'] = df['tot_time_oran'] / df['total_time'] * 100
df['percent_blue'] = df['tot_time_blue'] / df['total_time'] * 100
df['percent_neutral'] = df['tot_time_neutral'] / df['total_time'] * 100

# Plot the DataFrame
df[['percent_oran', 'percent_blue', 'percent_neutral']].plot.barh(stacked=True, figsize=(10, 20), color=['darkorange', 'royalblue', 'darkgrey'])
plt.xlabel('Percentage of the time')
plt.ylabel('Fly')

plt.show()
```

These barplots displays the percentage of time flies, accustomed to the two types of food or not habituated, spent with each food objects relative to the total time spent interacting with food.

```python
# Filter the DataFrame
df_orange = df[df['habituation_cond'] == 'orange'].copy()

# Calculate the total time for each row
df_orange['total_time_food'] = df_orange['tot_time_oran'] + df_orange['tot_time_blue']

# Calculate the percentage of time spent on each type of object
df_orange['percent_oran'] = df_orange['tot_time_oran'] / df_orange['total_time_food'] * 100
df_orange['percent_blue'] = df_orange['tot_time_blue'] / df_orange['total_time_food'] * 100

# Sort the DataFrame by 'percent_oran'
df_orange = df_orange.sort_values('percent_oran')

# Plot the DataFrame
ax = df_orange[['percent_oran', 'percent_blue']].plot.barh(
    stacked=True,
    color=['darkorange', 'royalblue'],
    width=1.0
)

ax.set_xlabel('Percentage of the time')
ax.set_ylabel('Flies habituated to orange')

plt.show()
```

This scatterplot with the accompanying boxplot evaluates the efficiency of the starving condition. It displays the percentage of time spent on each object relative to the total time of the experiment.

```python
# Filter the DataFrame
df_starved = df[df['starving_cond'] == 'starved'].copy()
df_fed = df[df['starving_cond'] == 'fed'].copy()

# Calculate the total time for each row
df_starved.loc[:, 'total_time_color'] = df_starved['tot_time_oran'] + df_starved['tot_time_blue'] 
df_fed.loc[:, 'total_time_color'] = df_fed['tot_time_oran'] + df_fed['tot_time_blue'] 

# Calculate the percentage of time spent on each type of object
df_starved.loc[:, 'percent_color'] = df_starved['total_time_color'] / df_starved['total_time'] * 100
df_starved.loc[:, 'percent_neutral'] = df_starved['tot_time_neutral'] / df_starved['total_time'] * 100

df_fed.loc[:, 'percent_color'] = df_fed['total_time_color'] / df_fed['total_time'] * 100
df_fed.loc[:, 'percent_neutral'] = df_fed['tot_time_neutral'] / df_fed['total_time'] * 100


#  Define the color mapping
color_dict = {'starved': 'darkred', 'fed': 'darkgreen'}

# Create a new DataFrame by concatenating df_starved and df_fed
df_combined = pd.concat([df_starved, df_fed])

# Create a figure with one subplot
fig, ax = plt.subplots(figsize=(10, 5))

# Plot the scatterplot and boxplot for 'percent_color'
sns.scatterplot(data=df_combined, x='percent_color', y='starving_cond', hue='starving_cond', palette=color_dict, legend=False, marker='x', linewidth=2, ax=ax)
sns.boxplot(data=df_combined, x='percent_color', y='starving_cond', color='white', linewidth=1, fliersize=0, boxprops=dict(alpha=0.5), width=0.4, ax=ax)
ax.tick_params(axis='both', labelsize=16, rotation=45)  
ax.set_xlabel('Percentage of Time on Food-Objects', fontsize=14, weight='bold')
ax.set_ylabel('Starving Condition', fontsize=14, weight='bold')


plt.tight_layout()
plt.subplots_adjust(top=0.88)  
plt.show()
```

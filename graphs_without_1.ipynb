```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Load the data
df = pd.read_csv('flydata_without_1.csv')
```

These firsts plots show the raw data of the three habituation conditions, over the total time spent with the orange or the blue object, independent on the feeding state. On the table we extracted the median of each condition.

```python
# Define your color mapping
color_dict = {'orange': 'darkorange', 'blue': 'royalblue', 'unused': 'gray'}

# Create a figure with two subplots (one row, two columns)
fig, axs = plt.subplots(1, 2, figsize=(20, 10))

# Add a title to the figure
fig.suptitle('Total time on food-objects under different habituation conditions', fontsize=16)

# Plot the scatterplot and boxplot for 'tot_time_orange' on the second subplot
sns.scatterplot(data=df, x='tot_time_oran', y='habituation_cond', hue='habituation_cond', palette=color_dict, legend=False, marker='x',linewidth=2, ax=axs[1])
sns.boxplot(data=df, x='tot_time_oran', y='habituation_cond', color='white', linewidth=1, fliersize=0, boxprops=dict(alpha=0.5), width=0.4, ax=axs[1])
axs[1].set_xlabel('Total Time on Orange')
axs[1].set_ylabel('Habituation Condition')
axs[1].set_title('Total Time on Orange')

# Plot the scatterplot and boxplot for 'tot_time_blue' on the first subplot
sns.scatterplot(data=df, x='tot_time_blue', y='habituation_cond', hue='habituation_cond', palette=color_dict, legend=False, marker='x', linewidth=2, ax=axs[0])
sns.boxplot(data=df, x='tot_time_blue', y='habituation_cond', color='white', linewidth=1, fliersize=0, boxprops=dict(alpha=0.5), width=0.4, ax=axs[0])
axs[0].set_xlabel('Total Time on Blue')
axs[0].set_ylabel('Habituation Condition')
axs[0].set_title('Total Time on Blue')

# Show the plot
plt.tight_layout()
plt.subplots_adjust(top=0.92)  # Adjust the top padding to make room for the title
plt.show()
```

```python
# Calculate the median of 'tot_time_oran' for each 'habituation_cond'
medians = df.groupby('habituation_cond')['tot_time_blue'].median()

# Convert the Series to a DataFrame and reset the index
medians_df = medians.reset_index()

# Rename the columns
medians_df.columns = ['Habituation Condition', 'Median Total Time on Blue']

# Display the DataFrame
medians_df
```

```python
# Calculate the median of 'tot_time_oran' for each 'habituation_cond'
medians = df.groupby('habituation_cond')['tot_time_oran'].median()

# Convert the Series to a DataFrame and reset the index
medians_df = medians.reset_index()

# Rename the columns
medians_df.columns = ['Habituation Condition', 'Median Total Time on Orange']

# Display the DataFrame
medians_df
```

Percentage of the time spent with each object and in the neutral area, per each fly.

```python
# Calculate the total time for each row
df['total_time'] = df['tot_time_oran'] + df['tot_time_blue'] + df['tot_time_neutral']

# Calculate the percentage of time spent on each type of object
df['percent_oran'] = df['tot_time_oran'] / df['total_time'] * 100
df['percent_blue'] = df['tot_time_blue'] / df['total_time'] * 100
df['percent_neutral'] = df['tot_time_neutral'] / df['total_time'] * 100

# Plot the DataFrame
df[['percent_oran', 'percent_blue', 'percent_neutral']].plot.barh(stacked=True, figsize=(10, 20), color=['darkorange', 'royalblue', 'darkgrey'])
plt.xlabel('Percentage of the time')
plt.ylabel('Fly')
# Show the plot
plt.show()
```

Percentage of time spent with the two food-object for flies used to the differents type of food, over the total time spent interacting with food.

```python
# Filter the DataFrame
df_orange = df[df['habituation_cond'] == 'orange'].copy()

# Calculate the total time for each row
df_orange.loc[:, 'total_time_food'] = df_orange['tot_time_oran'] + df_orange['tot_time_blue'] 

# Calculate the percentage of time spent on each type of object
df_orange.loc[:, 'percent_oran'] = df_orange['tot_time_oran'] / df_orange['total_time_food'] * 100
df_orange.loc[:, 'percent_blue'] = df_orange['tot_time_blue'] / df_orange['total_time_food'] * 100

# Plot the DataFrame
df_orange[['percent_oran', 'percent_blue']].plot.barh(stacked=True, color=['darkorange', 'royalblue'])
plt.xlabel('Percentage of the time')
plt.ylabel('Flies habituated to orange')
# Show the plot
plt.show()
```

```python
# Filter the DataFrame
df_blue = df[df['habituation_cond'] == 'blue'].copy()

# Calculate the total time for each row
df_blue.loc[:, 'total_time_food'] = df_blue['tot_time_oran'] + df_blue['tot_time_blue'] 

# Calculate the percentage of time spent on each type of object
df_blue.loc[:, 'percent_oran'] = df_blue['tot_time_oran'] / df_blue['total_time_food'] * 100
df_blue.loc[:, 'percent_blue'] = df_blue['tot_time_blue'] / df_blue['total_time_food'] * 100

# Plot the DataFrame
df_blue[['percent_oran', 'percent_blue']].plot.barh(stacked=True, color=['darkorange', 'royalblue'])
plt.xlabel('Percentage of the time')
plt.ylabel('Flies habituated to blue')
# Show the plot
plt.show()
```

These graphs are to check the efficientcy of the starving condition over the flies that are fed.

```python
# Filter the DataFrame
df_starved = df[df['starving_cond'] == 'starved'].copy()
df_fed = df[df['starving_cond'] == 'fed'].copy()

# Calculate the total time for each row
df_starved.loc[:, 'total_time_color'] = df_starved['tot_time_oran'] + df_starved['tot_time_blue'] 
df_fed.loc[:, 'total_time_color'] = df_fed['tot_time_oran'] + df_fed['tot_time_blue'] 

# Calculate the percentage of time spent on each type of object
df_starved.loc[:, 'percent_color'] = df_starved['total_time_color'] / df_starved['total_time'] * 100
df_starved.loc[:, 'percent_neutral'] = df_starved['tot_time_neutral'] / df_starved['total_time'] * 100

df_fed.loc[:, 'percent_color'] = df_fed['total_time_color'] / df_fed['total_time'] * 100
df_fed.loc[:, 'percent_neutral'] = df_fed['tot_time_neutral'] / df_fed['total_time'] * 100

# Create a figure with two subplots (two rows, one column)
fig, axs = plt.subplots(2, 1, figsize=(10, 20))

# Add a title to the figure
fig.suptitle('Percentage of time spent on food for starving and fed flies', fontsize=16)

# Plot the DataFrame for starved condition
df_starved[['percent_color', 'percent_neutral']].plot.barh(stacked=True, color=['darkred', 'gray'], ax=axs[0])
axs[0].set_xlabel('Percentage of the time')
axs[0].set_ylabel('Starved flies')

# Plot the DataFrame for fed condition
df_fed[['percent_color', 'percent_neutral']].plot.barh(stacked=True, color=['darkgreen', 'gray'], ax=axs[1])
axs[1].set_xlabel('Percentage of the time')
axs[1].set_ylabel('Fed flies')

# Show the plot
plt.tight_layout()
plt.subplots_adjust(top=0.95)  # Adjust the top padding to make room for the title
plt.show()
```

```python
# Calculate the overall percentage spent on color food for starved condition
overall_percent_color_starved = df_starved['total_time_color'].sum() / df_starved['total_time'].sum() * 100
print(f'Overall percentage spent on color food for starved condition: {overall_percent_color_starved}%')

# Calculate the overall percentage spent on color food for fed condition
overall_percent_color_fed = df_fed['total_time_color'].sum() / df_fed['total_time'].sum() * 100
print(f'Overall percentage spent on color food for fed condition: {overall_percent_color_fed}%')
```

```python
# Define your color mapping
color_dict = {'starved': 'darkred', 'fed': 'darkgreen'}

# Create a new DataFrame by concatenating df_starved and df_fed
df_combined = pd.concat([df_starved, df_fed])

# Create a figure with one subplot
fig, ax = plt.subplots(figsize=(10, 5))

# Plot the scatterplot and boxplot for 'percent_color'
sns.scatterplot(data=df_combined, x='percent_color', y='starving_cond', hue='starving_cond', palette=color_dict, legend=False, marker='x', linewidth=2, ax=ax)
sns.boxplot(data=df_combined, x='percent_color', y='starving_cond', color='white', linewidth=1, fliersize=0, boxprops=dict(alpha=0.5), width=0.4, ax=ax)
ax.set_xlabel('Percentage of Time on Food-Objects')
ax.set_ylabel('Starving Condition')

# Show the plot
plt.tight_layout()
plt.subplots_adjust(top=0.88)  # Adjust the top padding to make room for the title
plt.show()
```
